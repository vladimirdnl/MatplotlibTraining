```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,10,100)
plt.plot(x, x**0.5, label = 'Square Root', linestyle = '--')
plt.plot(x, np.sin(x), label = 'Sine', linestyle = '-.')
plt.legend()
```

```python
x = np.linspace(0,10,100)
fig, ax = plt.subplots(1)
ax.plot(x, np.sin(x))
ax.scatter(x, np.cos(x), s = 0.8)
```

```python
#More than one legend
x = np.linspace(0,5,100)
fig1, ax1 = plt.subplots(1)
plots = []
plots += ax1.plot(x, np.sin(x))
plots += ax1.plot(x, np.cos(x)**2, marker = '.')
plots += ax1.plot(x, x**0.5)
plots += ax1.plot(x, x**0.1)
ax1.axis('equal')
ax1.legend(plots[:2], ['sin x', 'cos$^2$ x'], loc = 'center right', frameon = False)
from matplotlib.legend import Legend
leg = Legend(ax1, plots[2:], ['x$^{0.5}$', 'x$^{0.1}$'], loc = 'upper left', frameon = False)
ax1.add_artist(leg)
```

```python
#simpliest colorbar
x = np.linspace(0,10,1000)
Im = np.sin(x)* np.cos(x[:, np.newaxis])
print(Im.shape)
plt.imshow(Im, cmap = 'cubehelix') #cmap = 'gray', 'viridis', 'RdBu'
#plt.imshow(Im, cmap = plt.cm.get_cmap('cubehelix',10)) #colormap using range of num -> 10 colors
plt.colorbar()
```

```python
from sklearn.datasets import load_digits
digits = load_digits(n_class = 6)
fig, ax = plt.subplots(8,8, figsize = (6,6))
for i, axi in enumerate(ax.flat):
    axi.imshow(digits.images[i], cmap = 'binary')
    axi.set(xticks = [], yticks = [])
```
