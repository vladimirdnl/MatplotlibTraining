```python
%matplotlib notebook
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("csv's/data.csv")
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
overall_median = 57287
```

```python
plt.plot(ages, dev_salaries, color = '#888888', linestyle = '--', label = 'All Devs')
plt.plot(ages, py_salaries, label = 'Python')
plt.fill_between(ages, py_salaries,dev_salaries, alpha = 0.1, 
                 where = (py_salaries>dev_salaries), interpolate = True,
                color = 'blue', label = 'Above avg')
plt.fill_between(ages, py_salaries,dev_salaries, alpha = 0.1,
                 where = (py_salaries<=dev_salaries), interpolate = True,
                color = 'red', label = 'Below avg')
plt.legend()
plt.grid(True)
plt.title('Median Salary by age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.savefig('pictures/Median_salary_filled.png')
```
