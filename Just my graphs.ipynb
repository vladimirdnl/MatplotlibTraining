```python
import matplotlib as mpl
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
#right_triangle_visualisation
x = np.array(range(1, 501))
y = np.array(range(1, 501))
#creating dataset
datadots = pd.DataFrame([i,j] for j in y for i in x)
datadots.columns = ['x coordinate','y coordinate']
print(datadots)
#finding hypothenuse
sqc = np.sqrt(np.power(datadots['x coordinate'], 2) + np.power(datadots['y coordinate'],2))
#indexes in dataset for hypothenuse as a Natural number
indexesright = sqc[sqc == sqc//1].index
for i in indexesright:
    plt.plot(datadots['x coordinate'].loc[i], datadots['y coordinate'].loc[i], marker = '.', color  = 'b') # - dots
    #plt.plot([0, datadots['x coordinate'].loc[i], datadots['x coordinate'].loc[i],0],
             #[0,datadots['y coordinate'].loc[i], 0,0]) # - triangles
```
